```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent

user = UserAgent()
main_url = 'https://www.tripadvisor.in'
page = requests.get(main_url, headers={'user-agent': user.chrome})
soup = BeautifulSoup(page.content, 'lxml')

all_div = soup.find_all('div', class_='HbQoi _t z')
#print(all_div)

links = [] #initializing a list to store all the href tags 

if len(all_div) >=1 : #checking if there are more than 1 div tags with the same name
   third_div = all_div[2] #accessing the 3rd div tag 

   objects = third_div.find_all('li', class_= 'Mkrpq Fg I _u') #access all the li tags 

   for obj in objects:

        link = obj.a['href']  # Extract the href attribute
        links.append(link)   # Append the link to the list

# Print the extracted links
for link in links:
    print(link)


print("Printing all absolute links\n")
for link in links:
    print(main_url + link)
   
    print("Data from the links for the best places to visit in that particular city")
    selected_url = f"{main_url}{link}" #navigate to the second web page 

    page = requests.get(selected_url, headers={'user-agent': user.chrome})
    bsoup = BeautifulSoup(page.content, 'lxml')

    inner_div = bsoup.find_all('div', class_='HbQoi _t z') 

    inner_links = [] #initializing a second list to store all the text data which is scraped from the 2nd web page

    if len(inner_div) >= 1: #checking if there are more than 1 div tags with the same name
        first_div = inner_div[2] #accessing the 1st div tag

        inner_objects = first_div.find_all('div', class_='keSJi FGwzt ukgoS')

        for obj in inner_objects:
            text = obj.get_text()  # Extract the text content of the <div>
            inner_links.append(text) # Append the text to the inner_links

    print(f"Data from {selected_url}:")
    for lk in inner_links:
        print(lk) #Prinitng all the scarped text from all the links one by one
    print("\n")




```
